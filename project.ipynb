<h1>Demonstração dos Dados obtidos de SP</h1>

```python
df_SP2021 = df2021[df2021['nom_estado'] == 'SAO PAULO']
df_SP2021.head()
```

```python
media_por_dia_SP2021 = df_SP2021.groupby(df_SP2021['din_instante'])['val_geracao'].mean().reset_index()
media_por_dia_SP2021.head()
```

Verificando a distribuição ao longo de Janeiro da energia gerada em SP

```python
janeiro_SP2021 = media_por_dia_SP2021[media_por_dia_SP2021['din_instante'].dt.month == 1]
janeiro_SP2021.reset_index(drop=True)

plt.figure(figsize=(10, 5))
sns.lineplot(x='din_instante', y='val_geracao', data=janeiro_SP2021)

plt.title('Variação dos Valores Gerados em SP ao longo do Mês de Janeiro')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.grid(True)
plt.xticks(rotation=45)
```

```python
soma_por_dia_SP2021 =  df_SP2021.groupby([df_SP2021['din_instante'].dt.date, 'nom_usina'])['val_geracao'].sum().reset_index()
soma_por_dia_SP2021.head()
```

Verificando se algum dia não foi gerada energia no Estado de SP

```python
soma_por_dia_SP2021[soma_por_dia_SP2021['val_geracao'] == 0].head()
```

Retirando as usinas que não produziram em determinado dia, visto que pode ser algo relacionado a problemas nas mesmas

```python
soma_por_dia_SP2021 = soma_por_dia_SP2021[soma_por_dia_SP2021['val_geracao'] != 0]
soma_por_dia_SP2021.head()
```

```python
mediasomas_por_dia_SP2021 = soma_por_dia_SP2021.groupby(soma_por_dia_SP2021['din_instante'])['val_geracao'].mean().reset_index()
mediasomas_por_dia_SP2021.head()
```

Verificando a distribuição da energia gerada por dia em SP

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='din_instante', y='val_geracao', data=mediasomas_por_dia_SP2021)
plt.title('Variação das Medias Geradas por dia em Sao Paulo ')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.grid(True)
plt.xticks(rotation=45)
```

Função para definir a estação do ano de acordo com data

```python
def estacao_do_ano(data):
    dia = data.day
    mes = data.month

    if (mes == 12 and dia >= 21) or (mes in [1, 2]) or (mes == 3 and dia < 21):
        return 'Verão'
    elif (mes == 3 and dia >= 21) or (mes in [4, 5]) or (mes == 6 and dia < 21):
        return 'Outono'
    elif (mes == 6 and dia >= 21) or (mes in [7, 8]) or (mes == 9 and dia < 21):
        return 'Inverno'
    elif (mes == 9 and dia >= 21) or (mes in [10, 11]) or (mes == 12 and dia < 21):
        return 'Primavera'
```

Verificando distribuição gerada por cada estação

```python
mediasomas_por_dia_SP2021['Estacao'] = mediasomas_por_dia_SP2021['din_instante'].apply(estacao_do_ano)
media_por_estacao_SP2021 = mediasomas_por_dia_SP2021.groupby('Estacao')['val_geracao'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Estacao', y='val_geracao', data=media_por_estacao_SP2021, palette='viridis', hue='Estacao', legend=False)

# Adicionando título e rótulos aos eixos
plt.title('Soma dos Valores por Estação do Ano')
plt.xlabel('Estação do Ano')
plt.ylabel('Soma dos Valores')

# Exibindo o gráfico
plt.show()
```

Definindo as métricas a serem usadas no Modelo

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import numpy as np
from prophet.diagnostics import cross_validation, performance_metrics
from prophet import Prophet
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Definindo o limite de tempo para treino
split_date = pd.to_datetime('2023-01-01')

df = df_filled_corrected.copy()

#Tratando outliers
lower_limit = 350
upper_limit = 1000
df['val_geracao'] = df['val_geracao'].apply(lambda x: np.nan if x < lower_limit or x > upper_limit else x)
df['val_geracao'] = df['val_geracao'].interpolate(method='linear')
df['val_geracao'] = df['val_geracao'].bfill().ffill()
df['val_geracao'].ffill()

# Dividindo os dados de treino e teste
df_train = df[pd.to_datetime(df['din_instante']) < split_date]
df_test = df[(pd.to_datetime(df['din_instante']) >= split_date)]
train = df_train['val_geracao']
test = df_test['val_geracao']

df.to_csv('Data/dataframe.csv')
df
```

Testando utilizar o modelo ARIMA

```python
#Definindo métricas do ARIMA
model = ARIMA(train, order=(30, 1,0))
model_fit = model.fit()

# Fazendo previsões com o modelo ajustado
forecast = model_fit.forecast(steps=len(test))

# Calculando MAE, MSE, RMSE e MAPE
mae = mean_absolute_error(test, forecast)
mse = mean_squared_error(test, forecast)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(test, forecast)

plt.figure(figsize=(14, 7))
plt.plot(df_train['din_instante'], train, label='Treinamento')
plt.plot(df_test['din_instante'], test, label='Real')
plt.plot(df_test['din_instante'], forecast, label='Previsão Arima', color='red')
plt.title('Previsão de Geração de Energia Solar com Arima')
plt.xlabel('Data')
plt.ylabel('Geração de Energia Solar')
plt.legend()
metrics_text = (f"MAE: {mae:.2f}\n"
                f"MSE: {mse:.2f}\n"
                f"RMSE: {rmse:.2f}\n"
                f"MAPE: {mape:.2%}")
plt.text(0.02, 0.95, metrics_text, transform=plt.gca().transAxes,
        fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))
plt.show()
```

Utilizando modelo SARIMA

```python
# Ajustar o modelo SARIMA
model_sarima = SARIMAX(train, order=(5, 1, 0), seasonal_order=(1, 1, 1, 10))
model_sarima_fit = model_sarima.fit(disp=False)

# Fazendo previsões
forecast_sarima = model_sarima_fit.forecast(steps=len(test))

# Calculando MAE, MSE, RMSE e MAPE para o SARIMA
mae_sarima = mean_absolute_error(test, forecast_sarima)
mse_sarima = mean_squared_error(test, forecast_sarima)
rmse_sarima = np.sqrt(mse_sarima)
mape_sarima = mean_absolute_percentage_error(test, forecast_sarima)

plt.figure(figsize=(14, 7))
plt.plot(df_train['din_instante'], train, label='Treinamento')
plt.plot(df_test['din_instante'], test, label='Real')
plt.plot(df_test['din_instante'], forecast_sarima, label='Previsão SARIMA', color='red')
plt.title('Previsão de Geração de Energia Solar com SARIMA')
plt.xlabel('Data')
plt.ylabel('Geração de Energia Solar')
plt.legend()
metrics_text = (f"MAE: {mae_sarima:.2f}\n"
                f"MSE: {mse_sarima:.2f}\n"
                f"RMSE: {rmse_sarima:.2f}\n"
                f"MAPE: {mape_sarima:.2%}")
plt.text(0.02, 0.95, metrics_text, transform=plt.gca().transAxes,
        fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))
plt.show()
```

Criando variáveis dummy para a estação do ano para possívelmente usar no modelo

```python
dummies_train = pd.get_dummies(df_train['Estacao'], prefix='Estacao')
dummies_test = pd.get_dummies(df_test['Estacao'], prefix='Estacao')
dummies = pd.get_dummies(df['Estacao'], prefix='Estacao')

# Adicionar as variáveis dummy ao DataFrame do Prophet
df_prophet_train = pd.concat([df_train[['din_instante', 'val_geracao']], dummies_train], axis=1)

df_prophet_train.rename(columns={'din_instante': 'ds', 'val_geracao': 'y'}, inplace=True)
df_prophet_train['ds'] = pd.to_datetime(df_prophet_train['ds'])
df_prophet_test = pd.concat([df_test[['din_instante', 'val_geracao']], dummies_test], axis=1)
df_prophet_test.rename(columns={'din_instante': 'ds', 'val_geracao': 'y'}, inplace=True)
df_prophet_test['ds'] = pd.to_datetime(df_prophet_test['ds'])
```

Utilizando modelo Prophet

```python
# Criando e ajustando o modelo Prophet
model_prophet = Prophet(
    yearly_seasonality=True, 
    weekly_seasonality=False,
    daily_seasonality=True,
    changepoint_prior_scale=0.1,
    seasonality_prior_scale=10.0
)
model_prophet.add_seasonality(name='monthly', period=30.5, fourier_order=5)
model_prophet.fit(df_prophet_train)

#Fazendo previsões
future = model_prophet.make_future_dataframe(periods=len(test))
forecast_prophet = model_prophet.predict(future)

# Calculando erros para o modelo Prophet
mae_prophet = mean_absolute_error(test, forecast_prophet['yhat'].iloc[-len(test):])
mse_prophet = mean_squared_error(test, forecast_prophet['yhat'].iloc[-len(test):])
rmse_prophet = np.sqrt(mse_prophet)
mape_prophet = mean_absolute_percentage_error(test, forecast_prophet['yhat'].iloc[-len(test):])

plt.figure(figsize=(14, 7))
plt.plot(df_train['din_instante'], train, label='Treinamento')
plt.plot(df_test['din_instante'], test, label='Real')
plt.plot(df_test['din_instante'], forecast_prophet['yhat'].iloc[-len(test):], label='Previsão Prophet', color='red')
plt.fill_between(df_test['din_instante'], forecast_prophet['yhat'].iloc[-len(test):], alpha=0.1,  color='red')
plt.title('Previsão de Geração de Energia Solar com Prophet')
plt.xlabel('Data')
plt.ylabel('Geração de Energia Solar')
plt.legend()
metrics_text = (f"MAE: {mae_prophet:.2f}\n"
                f"MSE: {mse_prophet:.2f}\n"
                f"RMSE: {rmse_prophet:.2f}\n"
                f"MAPE: {mape_prophet:.2%}")
plt.text(0.02, 0.95, metrics_text, transform=plt.gca().transAxes,
        fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))
plt.show()

# Realizando a validação cruzada e avaliando desempenho
df_cv = cross_validation(model_prophet, 
                        initial='730 days', 
                        period='180 days', 
                        horizon='365 days')
df_performance = performance_metrics(df_cv)
print(df_performance.head())
```

```python
#Definindo o funcionamento do modelo
def seasonal_naive_forecast(series, seasonal_periods, forecast_periods):
    if len(series) < seasonal_periods:
        raise ValueError("O comprimento da série deve ser maior ou igual ao número de períodos sazonais.")
    
    #Copio os dados presentes na serie temporal passada para a nova
    forecast_index = pd.date_range(start=series.index[-1] + pd.DateOffset(1), periods=forecast_periods, freq=series.index.freq)
    forecast_series = pd.Series(index=forecast_index, dtype=float)

    #Adiciono os dados futuros previstos
    for i in range(forecast_periods):
        forecast_series.iloc[i] = series.iloc[-seasonal_periods + i % seasonal_periods]
    
    return forecast_series

def filter_by_year(series, years):
    return series[series.index.year.isin(years)]

# Defina a sazonalidade (por exemplo, 365 para um ano) e o número de dias que deseja prever
seasonal_periods = 365
forecast_periods = 600

train_seasonal_naive = df_train.copy()
train_seasonal_naive.set_index('din_instante', inplace=True)
series = train_seasonal_naive['val_geracao']

#Faço as previsões
forecast_series = seasonal_naive_forecast(series, seasonal_periods, forecast_periods)

# Filtrar pelos anos desejados, nesse caso os que não possuiamos no treino
year_to_filter = [2023, 2024]
filtered_series = filter_by_year(forecast_series, year_to_filter)

print(f"Série para os anos {year_to_filter}:")
print(filtered_series)
```
